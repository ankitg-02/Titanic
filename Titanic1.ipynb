```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train=pd.read_csv(r'/content/train.csv')
```

```python
pd.DataFrame(train)
```

```python
t1=train.describe(include='all')
t1
```

```python
train['Age'].replace('',t1['Age'][4],inplace=True)
train['Embarked'].replace('',t1['Embarked'][2],inplace=True)
```

```python
train.drop('Cabin',axis=1)
```

```python
for i in train.columns:
  print(i,':','\n',(train[i].unique()))
```

```python
train['Survived'].value_counts()
```

```python
mod_train={'Passenger ID':train['PassengerId'],
           'Survived':train['Survived']}
train_data=pd.DataFrame(mod_train).set_index(['Passenger ID','Survived'])
train_data
```

```python
count=[100,200,300,400,500,600,700,800,900]
sns.distplot(mod_train['Survived'],count,label='Survival')
plt.legend()
plt.title('Survival Analysis')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```
